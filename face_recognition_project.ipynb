```python
import numpy as np
import pandas as pd
import os
import cv2
def dist1(v1,v2):
    return np.sqrt(((v1-v2)**2).sum())


def knn(train,test,k):
    dist=[]
    num = train.shape[0]
    
    for i in range(num):
        #this will provide me with the features of image
        ix = train[i, :-1]
        #this will provide me with the label of image
        iy = train[i,-1]
        
        d = dist1(ix,test)
        dist.append([d,iy])
        
        dk = sorted(dist,key = lambda x: x[0])[:k]
        labels = np.array(dk)[:, -1]
        pred = np.unique(labels,return_counts=True)
        index = np.argmax(pred[1]);
        return pred[0][index]
        
        
    
dataset_path = './Untitled Folder/'
facedata = []
labels = []

mapping = {}
classId = 0
    
## load your training data first....before just capturing the image......
for fx in os.listdir(dataset_path):
        if fx.endswith('.npy'):
            mapping[classId]  = fx[:-4]
            data_item = np.load(dataset_path+fx)
            facedata.append(data_item)
            
            label = classId*np.ones((data_item.shape[0],))
            classId += 1
            labels.append(label)

labels = np.concatenate(labels,axis = 0).reshape((-1,1))
facedata = np.concatenate(facedata,axis =0)

trainset = np.concatenate((facedata,labels),axis = 1)
### this is how you are going to train your dataset......
# __________________________________________________________________________________________________

## now i have to capture my camera

cap = cv2.VideoCapture(0)
face_classifier = cv2.CascadeClassifier('front_face.xml')

while True:
    ret,frame = cap.read()
    
    if ret==False:
         continue
    
    faces = face_classifier.detectMultiScale(frame,1.3,5)
    if(len(faces)==0):
        continue
         
    
    for face in faces:
        x,y,w,h = face
        
        offset = 10
        face_section = frame[y-offset:y+h+offset,x-offset:x+w+offset]
        face_section = cv2.resize(face_section,(100,100))
        ## this will give me the prediction.....of the class
        
        out = knn(trainset,face_section.flatten(),5)
        print(out)
        
        name = mapping[int(out)]
        cv2.putText(frame, name , (x+38,y+230), cv2.FONT_HERSHEY_SIMPLEX, 1 ,(255,0,255), 2, cv2.LINE_AA)
        cv2.rectangle(frame,(x,y),(x+w,y+h),(100,0,0),2)
           
    
    cv2.imshow("frame",frame)
    
    key_pressed = cv2.waitKey(1) & 0xFF
    
    if key_pressed== ord('q'):
        break


cap.release()
cv2.destroyAllWindows()
    
    
```
